```python
from scipy.stats import norm

#counting of people at each age interval shown in age_labels
age = [16, 23, 347 , 428 , 334 , 398 , 923 , 1015 , 216 , 11]
age_labels = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99']

#new vector 
vec = []

#indexing the values of ages
index = 0
for a in age:
    for i in range(a):
        vec.append(index)
    index = index + 1
    
mean = np.mean(vec)
deviation = np.std(vec)

#plot distribution graph
plt.title("Age Distribution on Diamond Pricess Cruise Ship", weight = "bold")
plt.xlabel('Age', weight = "bold")
plt.ylabel('Probability Density', weight = "bold")
#smooth graph by adding more datapoint
vec = np.arange(min(vec), max(vec), 0.1)
plt.plot(vec, norm.pdf(vec, mean, deviation))
plt.show()
```
